```python
import pandas as pd
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
df = pd.read_csv('train.csv')
```

```python
df.head()
```

# 1. Catagorical Data


## a. Countplot

```python
sns.countplot(df['Survived'])
df['Survived'].value_counts().plot(kind='bar')
```

```python
sns.countplot(df['Pclass'])
```

## b. Piechart

```python
df['Survived'].value_counts().plot(kind='pie',autopct= '%.2f')    # hor percentage "autopct= '%.2f'"
```

```python
df['Pclass'].value_counts().plot(kind='pie',autopct= '%.2f')    # hor percentage "autopct= '%.2f'"
```
