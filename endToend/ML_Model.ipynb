```python
df = df.iloc[0:,1:]     #[0: -rows,1: -column]
```

```python
X = df.iloc[:,0:2]
y = df.iloc[:,-1]
```

```python
X
```

```python
y
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test =  train_test_split(X,y,test_size = 0.1)        #0.1 means uses 10% data for test the model
```

```python
X_train
```

```python
y_train
```

```python
X_test
```

```python
from sklearn.preprocessing import StandardScaler
```

```python
scaler = StandardScaler()
```

```python
X_train = scaler.fit_transform(X_train)
```

```python
X_train                 #Now it becomed the array from the dataframe which has the values between 0 and 1
```

```python
X_test = scaler.transform(X_test)
```

```python
X_test
```

```python
#model Traing using Logistic Regression
```

```python
from sklearn.linear_model import LogisticRegression
```

```python
clf = LogisticRegression()
```

```python
clf.fit(X_train,y_train)
```

```python
#Evaluation
```

```python
y_predict = clf.predict(X_test)
```

```python
y_test
```

```python
from sklearn.metrics import accuracy_score
```

```python
accuracy_score(y_test, y_predict)
```

```python
#visualising the decision boundary
```

```python
from mlxtend.plotting import plot_decision_regions
```

```python
plot_decision_regions(X_train, y_train.values, clf=clf,legend=2)
```
